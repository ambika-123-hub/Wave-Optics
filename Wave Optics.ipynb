```python
# Interference
import numpy as np
import matplotlib.pyplot as plt

# Constants
wavelength = 0.5  # Wavelength of the light in micrometers (e.g., 0.5 micrometers for visible light)
k = 2 * np.pi / wavelength  # Wave number
distance = 100  # Distance between the two slits in micrometers
screen_distance = 1000  # Distance from the slits to the screen in micrometers
num_points = 1000  # Number of points on the screen

# Create the screen coordinates
x = np.linspace(-10, 10, num_points)  # Screen coordinates in micrometers

# Calculate the interference pattern
# Assume two sources (slits) at (-d/2, 0) and (d/2, 0)
d = distance
r1 = np.sqrt((x + d/2)**2 + screen_distance**2)
r2 = np.sqrt((x - d/2)**2 + screen_distance**2)

# Calculate the interference pattern
intensity = (np.cos(k * (r1 - r2)))**2

# Plot the interference pattern
plt.figure(figsize=(10, 6))
plt.plot(x, intensity, color='blue')
plt.title('Interference Pattern of Two Coherent Light Waves')
plt.xlabel('Position on Screen (micrometers)')
plt.ylabel('Intensity')
plt.grid(True)
plt.show()
```

# Diffraction

```python
# Diffraction pattern
import numpy as np
import matplotlib.pyplot as plt

# Constants
wavelength = 0.5e-6  # Wavelength of light in meters (e.g., 0.5 micrometers)
slit_width = 50e-6  # Slit width in meters (e.g., 50 micrometers)
screen_distance = 1  # Distance to screen in meters (e.g., 1 meter)
num_points = 1000  # Number of points on the screen
screen_range = 0.01  # Range of the screen in meters (e.g., +/- 1 cm)

# Create screen positions (x-coordinates)
x = np.linspace(-screen_range, screen_range, num_points)

# Calculate the diffraction pattern
# Fraunhofer single-slit diffraction equation: I = (I_0 * (sin(beta) / beta)^2)
beta = (np.pi * slit_width * x) / (wavelength * screen_distance)
intensity = (np.sin(beta) / beta)**2
intensity[beta == 0] = 1  # Handle division by zero at the center

# Plot the diffraction pattern
plt.figure(figsize=(10, 6))
plt.plot(x * 1e3, intensity, color='blue')  # x in mm
plt.title('Single-Slit Diffraction Pattern')
plt.xlabel('Position on Screen (mm)')
plt.ylabel('Intensity (a.u.)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
wavelength = 0.5e-6  # Wavelength of light in meters (e.g., 0.5 micrometers)
slit_width = 50e-6  # Width of each slit in meters (e.g., 50 micrometers)
slit_separation = 150e-6  # Distance between slits (center-to-center in meters)
n_slits = 10  # Number of slits
screen_distance = 1  # Distance to the screen in meters
num_points = 1000  # Number of points on the screen
screen_range = 0.01  # Range of the screen in meters (e.g., +/- 1 cm)

# Create screen positions (x-coordinates)
x = np.linspace(-screen_range, screen_range, num_points)

# Calculate the diffraction pattern
# Fraunhofer diffraction for n-slits
beta = (np.pi * slit_width * x) / (wavelength * screen_distance)
alpha = (np.pi * slit_separation * x) / (wavelength * screen_distance)

# Single-slit diffraction factor
diffraction = (np.sin(beta) / beta)**2
diffraction[beta == 0] = 1  # Handle division by zero at the center

# N-slit interference factor
interference = (np.sin(n_slits * alpha) / np.sin(alpha))**2
interference[alpha == 0] = n_slits**2  # Handle division by zero at the center

# Total intensity is the product of diffraction and interference
intensity = diffraction * interference

# Plot the diffraction pattern
plt.figure(figsize=(10, 6))
plt.plot(x * 1e3, intensity, color='blue')  # x in mm
plt.title(f'{n_slits}-Slit Diffraction Pattern')
plt.xlabel('Position on Screen (mm)')
plt.ylabel('Intensity (a.u.)')
plt.grid(True)
plt.show()
```

# Polarisation

```python
# Linear Polarisation
import numpy as np
import matplotlib.pyplot as plt

# Parameters for light wave
amplitude = 1       # Amplitude of the wave
frequency = 1       # Frequency of the wave
phase_difference = 0 # Phase difference between x and y components
time = np.linspace(0, 10, 1000)  # Time array

# Linear Polarization
# Electric field in the x and y direction for a linearly polarized light
Ex = amplitude * np.cos(2 * np.pi * frequency * time)
Ey = amplitude * np.cos(2 * np.pi * frequency * time + phase_difference)

# Plotting the electric field components and polarization
plt.figure(figsize=(8, 6))
plt.plot(Ex, Ey, label="Polarization", color='b')
plt.xlabel('Electric Field (E_x)')
plt.ylabel('Electric Field (E_y)')
plt.title('Linear Polarization of Light')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for light wave
amplitude = 1       # Amplitude of the wave
frequency = 1       # Frequency of the wave
phase_difference = np.pi / 2  # Phase difference of 90 degrees for circular polarization
time = np.linspace(0, 10, 1000)  # Time array

# Circular Polarization
# Electric field in the x and y direction for circularly polarized light
Ex = amplitude * np.cos(2 * np.pi * frequency * time)
Ey = amplitude * np.cos(2 * np.pi * frequency * time + phase_difference)

# Plotting the electric field components and polarization
plt.figure(figsize=(8, 6))
plt.plot(Ex, Ey, label="Circular Polarization", color='b')
plt.xlabel('Electric Field (E_x)')
plt.ylabel('Electric Field (E_y)')
plt.title('Circular Polarization of Light')
plt.grid(True)
plt.axis('equal')  # Set equal scaling for both axes
plt.legend()
plt.show()
```

```python
# Elliptical Polarisation
import numpy as np
import matplotlib.pyplot as plt

# Parameters for light wave
amplitude_x = 1       # Amplitude of the wave in the x direction
amplitude_y = 0.5     # Amplitude of the wave in the y direction
frequency = 1         # Frequency of the wave
phase_difference = np.pi / 3  # Phase difference for elliptical polarization
time = np.linspace(0, 10, 1000)  # Time array

# Elliptical Polarization
# Electric field in the x and y direction for elliptically polarized light
Ex = amplitude_x * np.cos(2 * np.pi * frequency * time)
Ey = amplitude_y * np.cos(2 * np.pi * frequency * time + phase_difference)

# Plotting the electric field components and polarization
plt.figure(figsize=(8, 6))
plt.plot(Ex, Ey, label="Elliptical Polarization", color='b')
plt.xlabel('Electric Field (E_x)')
plt.ylabel('Electric Field (E_y)')
plt.title('Elliptical Polarization of Light')
plt.grid(True)
plt.axis('equal')  # Set equal scaling for both axes
plt.legend()
plt.show()
```
